```python
#QUESTION 1
import pandas as pd
df =pd.read_csv("Used_Bikes.csv") 
df
data_filter =df[(df['city']=='Jaipur')&(df['brand']=='KTM')]
data_filter
print('Total number of bikes of brand KTM in jaipur city :',data_filter.shape)
df1 =data_filter['price']
print('Minimum price of KTM bike available in jaipur :',df1.min())
```

```python
#QUESTION 2
import pandas as pd
df =pd.read_csv("Used_Bikes.csv") 
df
data_filter =df[(df['city']=='Delhi')&(df['owner']=='First Owner')]
print('Total number of bikes in Delhi which are first hand are:',data_filter.shape)
df1 = data_filter['age']
print('Minimum age of a bike in Delhi which is first hand is :',df1.min(),'year')
print('Maximum age of a bike in Delhi which is first hand is :',df1.max(),'years')
```
